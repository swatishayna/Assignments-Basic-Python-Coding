```python
#Write a Python Program to Check if a Number is Positive, Negative or Zero?
num = float(input("Enter number here:   "))
if num > 0 :
    print(num, "is positive")
elif num < 0:
    print(num, "is negative")
else:
    print(num, "is zero")
```

```python
#Write a Python Program to Check if a Number is Odd or Even?
num = int(input("Enter number here:   "))
if num %2==0:
    print("even")
else:
    print("odd")
```

```python
#Write a Python Program to Check Leap Year?
year = input("Enter year here:   ")
if len(year)==4:
    try:
        int(year)
        if int(year)%4 == 0:
            print("Year is a leap year")
        else:
            print("year is not a leap year")
    except:
        print("Entered year is not valid try again with valid format For ex. 2021")
else:
    print("Entered year is not valid, try again with valid format For ex. 2021")
```

```python
#Write a Python Program to Check Prime Number?
n = int(input("Enter integer here:   "))
if n>2:
    for i in range(2,n):
        if n%i == 0:
            print(n,"number is composite")
            break
        else:
            print(n,"number is prime")
            break
elif n==1:
    print("1 is neither prime nor composite")
elif n==2:
    print(n, "is the smallest prime")
elif n==0:
    print(n, "zero")    
else:
    print("number is negative, enter any positive integer")
```

```python
#Write a Python Program to Print all Prime Numbers in an Interval of 1-10000
for n in range(1,10000):
    if n>2:
        for i in range(2,n):
            if n%i == 0:
                print(n,"number is composite")
                break
            else:
                print(n,"number is prime")
                break
    elif n==1:
        print("1 is neither prime nor composite")
    elif n==2:
        print(n, "is the smallest prime")
    elif n==0:
        print(n, "zero")  
    else:
        print("number is negative, enter any positive integer")
```
